```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator #resnet50 pipeline (data generators)
from tensorflow.keras.applications.resnet50 import preprocess_input

train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    brightness_range=[0.8, 1.2],
    zoom_range=0.1,
    horizontal_flip=False
)

val_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

train_generator = train_datagen.flow_from_directory(
    '/content/data/train',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
      '/content/data/val',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

```

```python
from tensorflow.keras.applications import ResNet50  #resnet50
from tensorflow.keras import layers, models
import tensorflow as tf

base_model = ResNet50(
    weights='imagenet',
    include_top=False,
    input_shape=(224, 224, 3)
)
base_model.trainable = False

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(train_generator.num_classes, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(   #resnet50 training
    train_generator,
    epochs=4,
    validation_data=val_generator
)
```

```python
import tensorflow as tf #finetuning resnet50

# Unfreeze part of ResNet50
base_model.trainable = True

for layer in base_model.layers[:-30]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_finetune = model.fit(
    train_generator,
    epochs=4,
    validation_data=val_generator
)



```

```python
import numpy as np #gradcam for inception v3
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
import tensorflow as tf

def display_gradcam(img_path, model, alpha=0.4):
    # 1. Prepare Image
    img = image.load_img(img_path, target_size=(299, 299))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = tf.keras.applications.inception_v3.preprocess_input(img_array)

    # 2. Extract the nested components

    seq_layer = next(l for l in model.layers if 'sequential' in l.name.lower())
    inception_backbone = next(l for l in seq_layer.layers if 'inception' in l.name.lower())

    # 3. Create a sub-model for the backbone that outputs the last CONV layer
    last_conv_layer_name = None
    for layer in reversed(inception_backbone.layers):
        if isinstance(layer, tf.keras.layers.Conv2D):
            last_conv_layer_name = layer.name
            break

    # Sub-model 1: Inputs -> Last Conv Layer of Inception
    # Sub-model 2: Inputs -> Final Inception Output
    backbone_grad_model = tf.keras.Model(
        inputs=inception_backbone.input,
        outputs=[inception_backbone.get_layer(last_conv_layer_name).output, inception_backbone.output]
    )

    # 4. Gradient Tape logic
    with tf.GradientTape() as tape:
        # Step A: Get conv features and backbone predictions
        last_conv_layer_output, backbone_output = backbone_grad_model(img_array)

        # Step B: Pass backbone output through the REST of the sequential layers

        x = backbone_output
        found_backbone = False
        for layer in seq_layer.layers:
            if found_backbone: # Only apply layers AFTER the inception backbone
                x = layer(x)
            if 'inception' in layer.name.lower():
                found_backbone = True

        predictions = x
        class_idx = tf.argmax(predictions[0])
        loss = predictions[:, class_idx]

    # 5. Compute Gradients
    grads = tape.gradient(loss, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    # 6. Build Heatmap
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    # Normalize
    heatmap = tf.maximum(heatmap, 0) / (tf.math.reduce_max(heatmap) + 1e-10)
    heatmap = heatmap.numpy()

    # 7. Visualize
    img_cv = cv2.imread(img_path)
    img_cv = cv2.resize(img_cv, (299, 299))
    heatmap_res = cv2.resize(heatmap, (299, 299))
    heatmap_res = np.uint8(255 * heatmap_res)
    heatmap_res = cv2.applyColorMap(heatmap_res, cv2.COLORMAP_JET)

    superimposed_img = cv2.addWeighted(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB), 0.6,
                                       cv2.cvtColor(heatmap_res, cv2.COLOR_BGR2RGB), 0.4, 0)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB))
    plt.title("Original")
    plt.axis("off")
    plt.subplot(1, 2, 2)
    plt.imshow(superimposed_img)
    plt.title(f"Grad-CAM Class: {class_idx.numpy()}")
    plt.axis("off")
    plt.show()

# Run it
display_gradcam(
    "/content/drive/MyDrive/ai data set/archive (2)/dataset/B-samples/100.jpg",
    model_inc_f
)
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
import tensorflow as tf

def display_gradcam_efficientnet(img_path, model, alpha=0.4):
    # 1. Prepare Image (EfficientNet usually takes 224x224 for B0)
    # Check your model's input shape; if it's B0, use 224. If it's B5, use 456.
    input_size = model.input_shape[1]
    img = image.load_img(img_path, target_size=(input_size, input_size))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)

    # EfficientNet has its own preprocessing (usually just rescaling/nothing)
    # If you used tf.keras.applications.efficientnet.preprocess_input during training:
    img_array = tf.keras.applications.efficientnet.preprocess_input(img_array)

    # 2. Extract nested components (Sequential -> EfficientNet)
    seq_layer = next(l for l in model.layers if 'sequential' in l.name.lower())
    eff_backbone = next(l for l in seq_layer.layers if 'efficientnet' in l.name.lower())

    # 3. Find the last Conv layer (EfficientNet uses 'top_conv' usually)
    last_conv_layer_name = None
    for layer in reversed(eff_backbone.layers):
        if isinstance(layer, tf.keras.layers.Conv2D):
            last_conv_layer_name = layer.name
            break

    # Sub-model for the backbone
    backbone_grad_model = tf.keras.Model(
        inputs=eff_backbone.input,
        outputs=[eff_backbone.get_layer(last_conv_layer_name).output, eff_backbone.output]
    )

    # 4. Gradient Tape logic
    with tf.GradientTape() as tape:
        # Step A: Get conv features and backbone predictions
        last_conv_layer_output, backbone_output = backbone_grad_model(img_array)

        # Step B: Manual Replay of the Top Head
        x = backbone_output
        found_backbone = False
        for layer in seq_layer.layers:
            if found_backbone:
                x = layer(x)
            if 'efficientnet' in layer.name.lower():
                found_backbone = True

        predictions = x
        class_idx = tf.argmax(predictions[0])
        loss = predictions[:, class_idx]

    # 5. Compute Gradients and Heatmap
    grads = tape.gradient(loss, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    # Normalize heatmap
    heatmap = tf.maximum(heatmap, 0) / (tf.math.reduce_max(heatmap) + 1e-10)
    heatmap = heatmap.numpy()

    # 6. Visualization
    img_cv = cv2.imread(img_path)
    img_cv = cv2.resize(img_cv, (input_size, input_size))
    heatmap_res = cv2.resize(heatmap, (input_size, input_size))
    heatmap_res = np.uint8(255 * heatmap_res)
    heatmap_res = cv2.applyColorMap(heatmap_res, cv2.COLORMAP_JET)

    superimposed_img = cv2.addWeighted(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB), 0.6,
                                       cv2.cvtColor(heatmap_res, cv2.COLOR_BGR2RGB), 0.4, 0)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB))
    plt.title("Original")
    plt.axis("off")
    plt.subplot(1, 2, 2)
    plt.imshow(superimposed_img)
    plt.title(f"EfficientNet Grad-CAM (Class {class_idx.numpy()})")
    plt.axis("off")
    plt.show()

# Run the function for your EfficientNet model (ensure the variable name matches)
display_gradcam_efficientnet(
    "/content/drive/MyDrive/ai data set/archive (2)/dataset/B-samples/100.jpg",
    model_eff_f
)
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image

def display_gradcam_resnet(img_path, model, alpha=0.4):
    # 1. Prepare Image
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)

    # ResNet50 specific preprocessing (BGR conversion and centering)
    img_array = tf.keras.applications.resnet50.preprocess_input(img_array)

    # 2. Extract nested components
    # Finds the sequential container and the resnet backbone
    try:
        seq_layer = next(l for l in model.layers if 'sequential' in l.name.lower())
        res_backbone = next(l for l in seq_layer.layers if 'resnet' in l.name.lower())
    except StopIteration:
        print("Error: Could not find 'sequential' or 'resnet' layers. Check model.summary()")
        return

    # 3. Find the last Conv2D layer in the ResNet backbone
    last_conv_layer_name = None
    for layer in reversed(res_backbone.layers):
        if isinstance(layer, tf.keras.layers.Conv2D):
            last_conv_layer_name = layer.name
            break

    # Sub-model for the ResNet backbone
    backbone_grad_model = tf.keras.Model(
        inputs=res_backbone.input,
        outputs=[res_backbone.get_layer(last_conv_layer_name).output, res_backbone.output]
    )

    # 4. Gradient Tape logic
    with tf.GradientTape() as tape:
        last_conv_layer_output, backbone_output = backbone_grad_model(img_array)

        # Manual Replay of the Top Head (your fine-tuned Dense layers)
        x = backbone_output
        found_backbone = False
        for layer in seq_layer.layers:
            if found_backbone:
                x = layer(x)
            if 'resnet' in layer.name.lower():
                found_backbone = True

        predictions = x
        class_idx = tf.argmax(predictions[0])
        loss = predictions[:, class_idx]

    # 5. Compute Gradients
    grads = tape.gradient(loss, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    # 6. Build Heatmap
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    # Normalize heatmap
    heatmap = tf.maximum(heatmap, 0) / (tf.math.reduce_max(heatmap) + 1e-10)
    heatmap = heatmap.numpy()

    # 7. Visualization
    img_cv = cv2.imread(img_path)
    img_cv = cv2.resize(img_cv, (224, 224))
    heatmap_res = cv2.resize(heatmap, (224, 224))
    heatmap_res = np.uint8(255 * heatmap_res)
    heatmap_res = cv2.applyColorMap(heatmap_res, cv2.COLORMAP_JET)

    # Convert BGR to RGB for correct plotting
    img_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
    heatmap_rgb = cv2.cvtColor(heatmap_res, cv2.COLOR_BGR2RGB)
    overlay = cv2.addWeighted(img_rgb, 0.6, heatmap_rgb, 0.4, 0)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(img_rgb)
    plt.title("Original ASL Image")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(overlay)
    plt.title(f"ResNet50 Grad-CAM (Class {class_idx.numpy()})")
    plt.axis("off")
    plt.show()

# Run the function
display_gradcam_resnet(
    "/content/drive/MyDrive/ai data set/archive (2)/dataset/B-samples/100.jpg",
    model_res_f
)
```
